```python
import pandas as pd
sylvey_annotated = pd.read_csv('Processed_Headings/sylvey_annotated_300.csv')
joe_annotated = pd.read_csv('Processed_Headings/joe_annotated_500.csv')

```

```python
sylvey_annotated
```

```python
joe_annotated
```

```python
common_columns = sylvey_annotated.columns.intersection(joe_annotated.columns)
sylvey_annotated = sylvey_annotated[common_columns]
joe_annotated = joe_annotated[common_columns]

# 合併
combined_annotated = pd.concat([sylvey_annotated, joe_annotated], ignore_index=True)
combined_annotated
```

```python
from sklearn.model_selection import train_test_split

# 假設 combined_df 是你剛剛合併好的 DataFrame
train_annotated, test_annotated = train_test_split(
    combined_annotated,
    test_size=0.2,
    random_state=42,
    shuffle=True
)
```

```python
train_annotated.to_csv("data/train_annotated.csv", index=False)
test_annotated.to_csv("data/test_annotated.csv", index=False)
```
