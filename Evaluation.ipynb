```python
import urllib.request as urllib2
import pickle
from collections import defaultdict

target_url = 'https://lhncbc.nlm.nih.gov/ii/areas/structured-abstracts/downloads/Structured-Abstracts-Labels-102615.txt'
normalized_sections = defaultdict(list)
with open('structured_headings.txt', 'w') as file:
    for line in urllib2.urlopen(target_url):
        clean_line = line.decode("utf-8").split('|')
        normalized_sections[clean_line[1].lower()].append(clean_line[0].lower())

with open('structured_abstract_sections.pkl', 'wb') as file:
    pickle.dump(normalized_sections, file)

print(normalized_sections)
print(list(normalized_sections))
```

```python
section_name_to_label = {
    '': 0,
    'methods': 1, 
    'background': 2, 
    'results': 3, 
    'conclusions': 4, 
    'objective': 5, 
}
```

```python
import pandas as pd

model = "angleloss"
n_neighbors = 5000
total_datapoints = f'{300000}'
file_name = f"clustering/{model}/clustering_labels_{n_neighbors}_{total_datapoints}.xlsx"

df = pd.read_excel(file_name)

from collections import Counter

for column in df.columns:
    cluster_content = df[column].dropna().tolist()
    mapping_section = [anonymous_mapping_dict[x.lower()] for x in cluster_content]
    
    counter = Counter(mapping_section)
    total = sum(counter.values())
    percentage = {key: (value / total) * 100 for key, value in counter.items()}
    print(percentage)

    counter = Counter(cluster_content)
    total = sum(counter.values())
    percentage = {key: (value / total) * 100 for key, value in counter.items()}
    print(percentage)
    print()

        
```

```python
from sklearn.metrics import adjusted_rand_score
true_labels = []
pred_labels = [] 
for i, column in enumerate(df.columns):
    
    cluster_content = df[column].dropna().tolist()
    mapping_section = [section_name_to_label[anonymous_mapping_dict[x.lower()]] for x in cluster_content]

    true_labels += mapping_section
    pred_labels += [i] * len(cluster_content)


ari_score = adjusted_rand_score(true_labels, pred_labels)
print("Adjusted Rand Index (ARI):", ari_score)
```
