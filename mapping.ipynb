```python
import pandas as pd

all_data = pd.read_csv('Processed_Headings/combined.csv')
all_data.head()
```

```python
all_data['sec-title'] = all_data.apply(lambda row: row['xmlPath'].rsplit(' > ', 1)[-1].strip(), axis=1)
all_data.head()
```

```python
all_data['sec-type'].unique().tolist()
```

```python
import pickle
with open('anonymous_mapping_dict.pkl', 'rb') as f:
    anonymous_mapping_dict = pickle.load(f)
anonymous_mapping_dict
```

```python
not_mapped = []
for sec_type in all_data['sec-type'].str.lower().dropna().unique().tolist():
    if sec_type not in anonymous_mapping_dict:
        not_mapped.append(sec_type)

print(len(not_mapped), not_mapped)
```

```python
len(all_data['sec-type'].str.lower().dropna().unique().tolist())
```

```python
not_mapped_title = []
for title in all_data['sec-title'].str.lower().dropna().unique().tolist():
    if title not in anonymous_mapping_dict:
        not_mapped_title.append(title)

print(len(not_mapped_title))
```

```python
len(all_data['sec-title'].str.lower().dropna().unique().tolist())
```

```python
not_mapped_title_type = []
for i, row in all_data.iterrows():
    if pd.isna(row['sec-type']) == False and row['sec-title'] not in anonymous_mapping_dict:
        not_mapped_title_type.append((row['sec-title'], row['sec-type']))

print(len(not_mapped_title_type))
```

```python
unique_values = all_data['sec-type'].str.lower().dropna().unique().tolist()
with open('sec-type_unique_values.pkl', 'wb') as f:
    pickle.dump(unique_values, f)
```
